```python
import datetime

pisos = 10
departamentos_por_piso = 4
departamentos_disponibles = [[''] * departamentos_por_piso for _ in range(pisos)]
departamentos_vendidos = [[''] * departamentos_por_piso for _ in range(pisos)]
precios_venta = {
    'A': 3800,
    'B': 3000,
    'C': 2800,
    'D': 3500
}
compradores = []



for piso in range(pisos):
    for depto in range(departamentos_por_piso):
        departamentos_disponibles[piso][depto] = True



def mostrar_departamentos_disponibles():
    print(" __________________________________")
    print("|\t\033[31m\033[1mTipo de Departamento \033[0m \033[0m     |")
    print("|__________________________________|")
    for piso in range(pisos):
        print(f"|Piso|  {piso+1}  |", end="  ")
        for depto in range(departamentos_por_piso):
            tipo = chr(ord('A') + depto)
            if departamentos_disponibles[piso][depto]:
                estado = ' '
            elif departamentos_vendidos[piso][depto]:
                estado = 'X'
            print(f"{tipo}:{estado} ", end="")
        print("   |")

def comprar_departamento():
    mostrar_departamentos_disponibles()
    piso = int(input("\nIngresa el número de piso: ")) - 1
    tipo = input("Ingresa el tipo de departamento (A, B, C, D): ").upper()
    if tipo not in precios_venta:
        print("Tipo de departamento inválido.")
        return

    if not departamentos_disponibles[piso][ord(tipo) - ord('A')]:
        print("El departamento seleccionado no está disponible.")
        return

    precio = precios_venta[tipo]
    estado = "X"
    departamentos_vendidos[piso][ord(tipo) - ord('A')] = True

    run = input("Ingresa el RUN del comprador(sin guion,sin puntos y sin digito vereficador): ")
    if not run.isdigit():
        print("RUN inválido.")
        return

    compradores.append((run, estado))
    departamentos_disponibles[piso][ord(tipo) - ord('A')] = False

    print("\n\033[32mLa operación se realizo con éxito.\033[0m")

def mostrar_listado_compradores():
    if not compradores:
        print("No hay compradores registrados.")
        return

    compradores_ordenados = sorted(compradores, key=lambda x: x[0])
    print("Listado de compradores:")
    for comprador in compradores_ordenados:
        print(f"RUN: {comprador[0]} - Estado: Vendido")


def mostrar_ganancias_totales():
    tipos_departamento = ['A', 'B', 'C', 'D']
    total_ventas = 0
    total_ventas_realizadas = 0

    print("          \033[31m\t\033[1mGANANCIAS TOTALES\033[0m\033[0m                      ")
    print("----------------------------------------------------")
    print(" Tipo Departamento | Ventas Realizadas | Total en UF")
    print("----------------------------------------------------")

    for tipo in tipos_departamento:
        valor_departamento = precios_venta[tipo]
        ventas_realizadas = sum([1 for piso in range(pisos) if departamentos_vendidos[piso][ord(tipo) - ord('A')]])
        ventas_tipo = valor_departamento * ventas_realizadas
        total_ventas += ventas_tipo
        total_ventas_realizadas += ventas_realizadas
        print(f" Tipo {tipo} {valor_departamento:<8}UF | {ventas_realizadas:<16}  |  {ventas_tipo:<12}")

    print("----------------------------------------------------")
    print(f"Total              | \033[32m{total_ventas_realizadas:<16}\033[0m  |\033[32m  {total_ventas:<12}\033[0m")
    print("----------------------------------------------------")



def main():
    while True:
        print("______________________________________")
        print("|                                    |")
        print("|\t\033[33mInmobiliaria Casa Feliz\033[0m      |")
        print("|                                    |")
        print("|\tseleccione una opción:       |")
        print("|____________________________________|")
        print("|                                    |")
        print("|\033[31m1.\033[0m Comprar departamento             |")
        print("|\033[31m2.\033[0m Mostrar departamentos disponibles|")
        print("|\033[31m3.\033[0m Ver listado de compradores       |")
        print("|\033[31m4.\033[0m Mostrar ganancias totales        |")
        print("|\033[31m5.\033[0m Salir                            |")


        opcion = input(" \n Opción:    ")
        if opcion == '1':
            comprar_departamento()
        elif opcion == '2':
            mostrar_departamentos_disponibles()
        elif opcion == '3':
            mostrar_listado_compradores()
        elif opcion == '4':
            mostrar_ganancias_totales()
        elif opcion == '5':
            fecha_actual = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            print(f" \033[1m\tGracias por usar nuestros sistemas\033[0m  \n\033[31m\033[1m CREADO POR FREDDY NEILAF GONZALEZ  {fecha_actual}.\033[0m \033[0m")
            break
        else:
            print("Opción inválida.")

main()
```
